```python
import numpy as np
import matplotlib.pyplot as plt

labels = ["Caso Pequeo", "Caso Medio", "Caso Grande"]

roi_r = np.array([323.3695, 386.6144, 326.3286])
roi_a = np.array([323.3695, 326.3286, 326.3286])

roi_r_plot = np.append(roi_r, roi_r[0])
roi_a_plot = np.append(roi_a, roi_a[0])

angles_deg = [30, 150, 270]
angles = np.radians(angles_deg)
angles = np.append(angles, angles[0])

fig = plt.figure(figsize=(6, 6))
ax = plt.subplot(111, polar=True)

line_r, = ax.plot(angles, roi_r_plot)
ax.fill(angles, roi_r_plot, alpha=0.25)

line_a, = ax.plot(angles, roi_a_plot)
ax.fill(angles, roi_a_plot, alpha=0.25)

ax.set_thetagrids(angles_deg, labels)
ax.set_title("Comparacin de ROI R* vs A*", pad=20)
ax.grid(True)
ax.set_rlabel_position(0)

plt.legend(
    [line_r, line_a],
    ["R*", "A*"],
    loc="upper right",
    bbox_to_anchor=(1.3, 1.1)
)

plt.show()
```
